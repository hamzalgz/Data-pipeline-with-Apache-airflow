```python
def add_student(prenom,nom,email):
    with open("/home/imade/notebook/config.json") as f:
        config = json.load(f)
    print("Config:",config)
    mydb = psycopg2.connect(database=config["database"], user=config["user"], password=config["password"], host=config["host"], port=config["port"], options=f'-c search_path=biabsence')
    mycursor = mydb.cursor()
    mycursor.execute(f"SELECT * FROM students_db.students where email ='{email}'")
    myresult = mycursor.fetchall()
    if len(myresult) != 0:
        return myresult[0][0]
    if len(myresult) == 0:
        sql = "INSERT INTO students_db.students (first_name,last_name,email) VALUES (%s,%s,%s)"
        val = (prenom,nom,email)
        mycursor.execute(sql, val)
        mydb.commit()
        # print(mycursor.rowcount, "record inserted.")
        mycursor.execute(f"SELECT * FROM students_db.students where email ='{email}'")
        myresult = mycursor.fetchall()
        return myresult[0][0]
```

```python
list_students = pd.read_excel(f"{input_file}",engine="openpyxl")
nom=[]
prenom=[]
email=[]
for index, row in list_students.iterrows():
    Nom_=row['Nom']
    prenom_ = Nom_.split(" ")[0]
    nom_ = Nom_.replace(prenom_+" ","")
    email_=f"{prenom_[0]}.{nom_.split(' ')[0]}@insea.ac.ma"
    
    nom.append(nom_)
    prenom.append(prenom_)
    email.append(email_)
    
list_students["prenom"]=prenom
list_students["nom"] = nom
list_students["email"]=email
del list_students['Nom']
list_students
```

```python
for index, row in list_students.iterrows():
    nom=row['nom']
    prenom=row['prenom']
    email=row['email']
    id_student=add_student(prenom,nom,email)
    print(id_student)
```
